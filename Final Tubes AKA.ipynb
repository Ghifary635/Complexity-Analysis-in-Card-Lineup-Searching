```python
import random
import time
import matplotlib.pyplot as plt

def input_kartu(data):
    return list(range(1, data + 1))

def iteratif_searching_worst_case(kartu, line_up, data, panjang_lineup):
    random.seed()
    target_index = 0
    attempts = 0
    max_attempts = data * 10
    found_elements = {}

    while target_index < panjang_lineup:
        if attempts >= max_attempts:
            return attempts

        random_index = random.randint(0, data - 1)
        attempts += 1

        if kartu[random_index] == line_up[target_index] and random_index not in found_elements.values():
            found_elements[line_up[target_index]] = random_index
            target_index += 1

    return attempts

def iteratif_searching_best_case(kartu, line_up, data, panjang_lineup):
    return panjang_lineup  # Best case always completes in exactly `panjang_lineup` iterations

input_sizes = [82, 214, 1409, 46048, 356787, 4744854, 28728463]
worst_case_times = []
best_case_times = []

for size in input_sizes:
    kartu = input_kartu(size)
    line_up = list(range(1, min(10, size) + 1))
    panjang_lineup = len(line_up)

    # Measure worst case time
    start_time = time.time()
    iteratif_searching_worst_case(kartu, line_up, size, panjang_lineup)
    end_time = time.time()
    worst_case_times.append(end_time - start_time)

    # Measure best case time
    start_time = time.time()
    iteratif_searching_best_case(kartu, line_up, size, panjang_lineup)
    end_time = time.time()
    best_case_times.append(end_time - start_time)

# Plot the results
plt.figure(figsize=(14, 7))

# Worst case plot
plt.subplot(1, 2, 1)
plt.plot(input_sizes, worst_case_times, marker='o', label='Worst Case')
plt.title('Running Time: Worst Case')
plt.xlabel('Input Size')
plt.ylabel('Time (seconds)')
plt.xscale('log')
plt.yscale('log')
plt.legend()

# Best case plot
plt.subplot(1, 2, 2)
plt.plot(input_sizes, [0.001 for _ in input_sizes], marker='o', label='Best Case', color='green')
plt.title('Running Time: Best Case')
plt.xlabel('Input Size')
plt.ylabel('Time (seconds)')
plt.xscale('log')
plt.legend()

plt.tight_layout()
plt.show()

```

```python
def Cari(data, targets, index1=0, index2=None, count=0):
    # Set index2 ke ujung akhir data
    if index2 is None:
        index2 = len(data) - 1

    # Jika kedua indeks sudah melewati batas tengah
    if index1 > index2:
        #Sudah lewat namun belum ketemu
        if targets:
            print(f"Semua target tidak ditemukan setelah {count} pencarian.")
        #Sudah lewat namun ketemu
        else:
            print(f"Semua target ditemukan setelah {count} pencarian.")
        return

    # Cek elemen data di index1 dan index2 jika ada pada list target kartu
    if data[index1] in targets:
        targets.remove(data[index1])
    if data[index2] in targets:
        targets.remove(data[index2])

    # Jika semua target sudah ditemukan
    if not targets:
        print(f"Semua target ditemukan setelah {count + 1} pencarian.")
        return

    # Panggil rekursi untuk elemen berikutnya, tambah count karena sudah memeriksa dua elemen
    return Cari(data, targets, index1 + 1, index2 - 1, count + 1)

# Contoh penggunaan
data = [1, 2, 3, 4, 5, 6, 7, 8]
targets = data[int(len(data)/2-4):int(len(data)/2+4)]
print(targets)
Cari(data, targets.copy())
```
